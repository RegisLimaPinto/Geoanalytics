```python
# ============================================================
# GeoPSI v4.0 - Parametric Susceptibility Index (PSI)
# Módulo Estatístico de Integração Geoespacial
# ============================================================

import numpy as np
import pandas as pd
from scipy.interpolate import griddata
from scipy.ndimage import gaussian_filter
import math

# ============================================================
# CONFIGURAÇÃO DO MODELO
# ============================================================

PSI_CONFIG = {
    "L0": 1.0,              # escala base do fator latente
    "lambda_decay": 0.35,   # decaimento espacial empírico
    "regularization": 0.10  # suavização do ajuste
}

# ============================================================
# UTILITÁRIOS
# ============================================================

def robust_normalize(data, p_low=5, p_high=95):
    lo = np.percentile(data[~np.isnan(data)], p_low)
    hi = np.percentile(data[~np.isnan(data)], p_high)

    if abs(hi - lo) < 1e-12:
        return np.zeros_like(data)

    norm = (data - lo) / (hi - lo)
    return np.clip(norm, 0, 1)


def create_grid_from_bbox(bbox, width, height):
    min_lon, min_lat, max_lon, max_lat = bbox
    lons = np.linspace(min_lon, max_lon, width)
    lats = np.linspace(min_lat, max_lat, height)
    return np.meshgrid(lons, lats)


def interpolate_points_to_grid(df, X, Y):
    return griddata(
        (df['lon'], df['lat']),
        df['value'],
        (X, Y),
        method='linear',
        fill_value=np.nanmedian(df['value'])
    )


# ============================================================
# CAMADA PSI (ESTATÍSTICA)
# ============================================================

def compute_shielding_index(density_grid, radiometric_grid=None):
    """
    Proxy estatístico de suscetibilidade local (Σ)
    """
    base = robust_normalize(density_grid)

    if radiometric_grid is not None:
        rad = robust_normalize(radiometric_grid)
        return 0.6 * base + 0.4 * rad

    return base


def compute_latent_field(sigma_grid, depth_grid):
    """
    Campo latente de correção espacial
    """
    L0 = PSI_CONFIG["L0"]
    lam = PSI_CONFIG["lambda_decay"]

    depth_norm = depth_grid / np.max(depth_grid)
    field = L0 * (1 - sigma_grid * np.exp(-lam * depth_norm))

    return np.clip(field, 0.01, 1.0)


def compute_shielding_gradient(sigma_grid):
    grad_y, grad_x = np.gradient(np.nan_to_num(sigma_grid))
    grad = np.sqrt(grad_x**2 + grad_y**2)

    if np.max(grad) > 0:
        grad = grad / np.max(grad)

    return grad


def psi_adjust_score(base_score, sigma, field, gradient):
    """
    Ajuste não linear do score
    """
    shielding_factor = 0.5 + 0.5 * sigma
    field_factor = 1.5 - 0.5 * field
    gradient_factor = 1.0 + gradient

    combined = (
        shielding_factor * 0.4 +
        field_factor * 0.3 +
        gradient_factor * 0.3
    )

    adjusted = base_score * np.tanh(combined)

    return np.clip(adjusted, 0, 1)


# ============================================================
# PIPELINE PRINCIPAL
# ============================================================

def process_geospatial_project(config, layers):
    """
    Pipeline completo de análise
    """

    # GRID
    X, Y = create_grid_from_bbox(
        config['bbox'],
        config['grid_w'],
        config['grid_h']
    )

    # NORMALIZAÇÃO
    normalized = {
        k: robust_normalize(v)
        for k, v in layers.items()
    }

    # DENSIDADE PROXY
    density = 0.6 * normalized.get("MAG", 0) + 0.4 * normalized.get("GRAV", 0)

    # SIGMA (suscetibilidade)
    sigma = compute_shielding_index(density)

    # PROFUNDIDADE SIMULADA
    center_lon = (config['bbox'][0] + config['bbox'][2]) / 2
    center_lat = (config['bbox'][1] + config['bbox'][3]) / 2

    depth = np.sqrt((X - center_lon)**2 + (Y - center_lat)**2)

    # CAMPO LATENTE
    field = compute_latent_field(sigma, depth)

    # GRADIENTE
    gradient = compute_shielding_gradient(sigma)

    # SCORE BASE
    weights = config.get("weights", {})
    base_score = np.zeros_like(X)

    for layer, w in weights.items():
        if layer in normalized:
            base_score += normalized[layer] * w

    base_score = robust_normalize(base_score)

    # SCORE AJUSTADO
    final_score = np.zeros_like(base_score)

    for i in range(base_score.shape[0]):
        for j in range(base_score.shape[1]):
            final_score[i, j] = psi_adjust_score(
                base_score[i, j],
                sigma[i, j],
                field[i, j],
                gradient[i, j]
            )

    # RESULTADOS
    return {
        "BASELINE_SCORE": base_score,
        "PSI_ADJUSTED_SCORE": final_score,
        "SHIELDING_INDEX": sigma,
        "LATENT_FIELD": field,
        "SHIELDING_GRADIENT": gradient
    }


# ============================================================
# RANKING DE ALVOS
# ============================================================

def rank_targets(results, config):
    final = results["PSI_ADJUSTED_SCORE"]
    X, Y = create_grid_from_bbox(
        config['bbox'],
        config['grid_w'],
        config['grid_h']
    )

    rankings = []

    for t in config['targets']:
        ix = int((t['lon'] - config['bbox'][0]) /
                 (config['bbox'][2] - config['bbox'][0]) *
                 config['grid_w'])

        iy = int((t['lat'] - config['bbox'][1]) /
                 (config['bbox'][3] - config['bbox'][1]) *
                 config['grid_h'])

        score = float(final[iy, ix])

        rankings.append({
            "target": t['id'],
            "score": score
        })

    return sorted(rankings, key=lambda x: -x['score'])
```
